#### Define a function called check_price - To check the price every so and so minutes as required

```python
def check_price():
    
    URL ="https://www.moneycontrol.com/india/stockpricequote/aluminium/nationalaluminiumcompany/NAC"
    
    headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36"}
    
    page = requests.get(URL, headers=headers)
    
    soup1 = BeautifulSoup(page.content,"html.parser")
    
    soup2 = BeautifulSoup(soup1.prettify(),"html.parser")
    
    title = soup2.find("div", class_= 'inid_name').h1.text.split()
    title = "".join(title)
    
    price = soup2.find(id = 'nsecp', class_ = 'inprice1 nsecp').get_text().split()[0]
    price = float(price)

    low_price = soup2.find(id = 'sp_low', class_ = 'FL nseLP').get_text().split()[0]
    low_price = float(low_price)
    
    high_price = soup2.find(id = 'sp_high', class_ = 'FR nseHP').get_text().split()[0]
    high_price = float(high_price)
    
    present_day = soup2.find('p', class_ = 'nseasondate').get_text().split()
    present_day = " ".join(present_day)

    
    header = ['Title','CurrentPrice','DayRange_low','DayRange_high']
    data = [title, price, low_price, high_price]

    with open('National_Alum_dataset.csv','w', newline='',encoding ='UTF8') as file:
        write = csv.writer(file)
        write.writerow(header)
        write.writerow(data)
        
    with open('Nationalalum_dataset.csv','a+',newline='',encoding ='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)
        
    if float(price) < 71.0:
                  send_mail()  
            
    return title , price, low_price, high_price, present_day
 
```

#### while loop to check price every 5 seconds

```python
while True:
    check_price()
    time.sleep(5)
```

```python
while(True):
    check_price()
    time.sleep(3600)
```
